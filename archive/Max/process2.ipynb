```python
case_vacc["date"] = pd.to_datetime(case_vacc["date"], format='%Y/%m/%d')
```

```python
case_vacc.head(1)
```

```python
case_vacc = case_vacc.sort_values(by="date",ascending=False)
```

```python
case_vacc.head(5)
```

```python
case_vacc_list = case_vacc.values.tolist()
```

```python
import math
allc = {}
for country in case_vacc_list:
    name = country[0]
    try:
        value = math.log(int(country[2]), 2)
    except:
        continue
    try:
        vacc = math.log(int(country[12]), 2)
    except:
        continue
    if name not in allc:
        allc[name]={"name": name, "value": value, "tvalue":country[2], "vacc": vacc, "tvacc":country[12]}
        print("{",end='')
        print(f'name: "{name}", value: {value}, tvalue: {country[2]}, vacc: {vacc}, tvacc: {country[12]}', end='},\n')
```

```python
detailData = {}
for country in case_vacc_list:
    name = country[0]
    year = country[1]
    year = year.strftime("%Y-%m-%d")
    value = country[2]
    vacc = country[12]

    if name not in detailData:
        detailData[name] = [{"year": year, "value": value, "vacc": vacc}]
    else:
        if len(detailData[name]) < 15:
            detailData[name].append({"year": year, "value": value, "vacc": vacc})
```

```python
# print(detailData)
```

```python
for m in detailData:
    for d in range(0, len(detailData[m])-1):
        # print(detailData[c][d])
        # print(detailData[c][d-1])
        # print(detailData[m][d]['vacc'])
        a = detailData[m][d]['year']
        new_case = detailData[m][d]['value']-detailData[m][d+1]['value']
        new_vacc = detailData[m][d]['vacc']
        # new_vacc = detailData[m][d]["vacc"]-detailData[m][d+1]["vacc"]

        print("{",end='')
        print(f'name: "{m}", year: "{a}", case: {new_case}, type: "New Case"', end='},\n')
        print("{",end='')
        print(f'name: "{m}", year: "{a}", case: {new_vacc}, type: "Daily Vaccination"', end='},\n')
```
